```python
url_prepend = 'https://www.sec.gov/Archives/'
```

```python
cik_list = pd.read_excel('cik_list.xlsx')
```

```python
import re
```

```python
ana_section = []
for row_index, row in cik_list.iterrows():
    url = (url_prepend + row['SECFNAME'])
    response = urllib.request.urlopen(url)
    file = response.read().decode('utf-8').lower()
    sections = re.findall('item[^a-zA-Z\n]\d\. *(management\'s discussion and analysis(?:.)*?)item[^a-zA-Z\n]\d\.',file,flags=re.DOTALL)
    sections.append('')
    ana_section.append(max(sections,key=len))
    
    
```

```python
np.sum([len(x) for x in ana_section ])
```

```python
for x in ana_section:
    print(len(x))
    print('')
```
